```python
import polars as pl
```

```python
# 1. Lazily read the Parquet dataset. This scans the file metadata but doesn't load data.
# Polars will automatically discover and use all .parquet files in the directory.
path_to_read = "/media3/majumder/dataset/cepheids/hats/zubercal_vcep/"
lazy_df = pl.scan_parquet(f"{path_to_read}**/*.parquet")
# majumder/dataset/cepheids/hats/zubercal_vcep/dataset/Norder=0/Dir=0/Npix=0.parquet
```

```python
type(lazy_df)
```

```python
print("Starting computation...")
eager_df = lazy_df.collect()
print("Computation finished!")

# Now you can work with the result as a normal Polars DataFrame
print(eager_df.head())
print(f"DataFrame shape: {eager_df.shape}")
```
